```python
import os
import matplotlib.pyplot as plt
import cv2
from matplotlib.widgets import RectangleSelector
from autoXML import write_xml
from xml.etree import ElementTree as ET
```

```python
# global constants
img = None
tl_list = []
br_list = []
object_list = []

# constants
image_folder = 'Images'
savedir = 'autoXML_test'
obj = 'tray'
xml_folder = 'Annotations'
```

```python
for n, image_file in enumerate(os.scandir(image_folder)):
    img = image_file
    
    if '(0)' in image_file.path:
        continue
    
    path_xml = str(xml_folder + image_file.path[6:-4] + '.xml')
    print(image_file)
    with open(path_xml, 'r') as myfile:
          tree_string = myfile.read()
    tree = ET.fromstring(tree_string)    
    
    min_x = 99999999999
    max_x = -1
    
    min_y = 99999999999
    max_y = -1
    
    for group in tree:
        if group.tag == "object":
            if group.findtext("name") == "tray":
                for subgroup in group:
                    for polygon in subgroup:
                        if polygon.tag == "pt":
                            for coords in polygon:  
                                if coords.tag == 'x':
                                    min_x = min(min_x, int(coords.text))
                                    max_x = max(max_x, int(coords.text))
                                elif coords.tag == 'y':
                                    min_y = min(min_y, int(coords.text))
                                    max_y = max(max_y, int(coords.text))
    tl_list = [(min_x, min_y)]
    br_list = [(max_x, max_y)]
    object_list = ['tray']
    
    
    
    write_xml(image_folder, img, object_list, tl_list, br_list, savedir)
```
